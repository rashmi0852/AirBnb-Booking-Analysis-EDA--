## 3. ***Data Wrangling***

### Data Wrangling Code

````python
#problem statement-1
#What is the distribution of listings in different neighborhood_groups in terms of total count & average listing price?(Univariate + Bivariate Analysis)

# Group data by neighborhood_group and calculate the count and average price
grouped_data = airbnb.groupby('neighbourhood_group').agg({'id': 'count', 'price': 'mean'}).reset_index()
grouped_data.rename(columns={'id': 'property_count', 'price': 'avg_price'}, inplace=True)

# Sort by property_count in descending order
neighbourhood_group_count_price= round(grouped_data.sort_values(by='property_count', ascending=False))

#printing result
print (f"Property listing count & price of neighbourhood groups\n" + "---"*18)
print(neighbourhood_group_count_price)
````

````python
#problem statement-2
#What are the different types of room available? & How does the count & price vary across different room types?(Univariate + Bivariate Analysys)

# Get the counts of each room type
room_type_counts = airbnb['room_type'].value_counts()

#printing roomtypes count
print (f"Room type total count"+"\n" + "--"*12)
print(room_type_counts)

# Calculate the average cost of each room type
avg_price_by_room_type = round(airbnb.groupby('room_type')['price'].mean())


#printing average price of each roomtype
print (f"\nRoom type average price"+"\n" + "--"*12)
print(avg_price_by_room_type)
````

````python
#problem statement-3
# How does the count of different room types vary across various neighborhood groups in the  listings?(Bivariate Analysis)

# Group data by neighborhood_group and room_type, then calculate counts
roomtype_counts_by_neighbourhood_group = airbnb.groupby(['neighbourhood_group', 'room_type']).size().reset_index(name='count')

#printing result
print (f"\n Room type count by each neighbourhood group\n" + "--"*20)
print(roomtype_counts_by_neighbourhood_group)
````

````python
#problem statement-4
#How does the room count vary among  neighbourhoods in each neighbourhood groups?(Top 3 & bottom 3)(Bivariate + Bivariate Analysis)

# Group data by neighbourhood_group and neighbourhood, then calculate room count
room_count_by_neighbourhood = airbnb.groupby(['neighbourhood_group', 'neighbourhood'])['id'].count().reset_index()

# Sort the data by neighbourhood_group and room count in descending order
room_count_by_neighbourhood = room_count_by_neighbourhood.sort_values(by=['neighbourhood_group', 'id'], ascending=[True, False])

# Select top 3 neighbourhoods within each neighbourhood_group
top_3_neighbourhoods_count = room_count_by_neighbourhood.groupby('neighbourhood_group').head(3)

#printing result
print (f"Top 3 neighbourhood  by listing count of each neighbourhood group\n" + "--"*33)
print(top_3_neighbourhoods_count)

# Select top 3 neighbourhoods within each neighbourhood_group
Bottom_3_neighbourhoods_count = room_count_by_neighbourhood.groupby('neighbourhood_group').tail(3)


#printing result
print (f"\nBottom 3 neighbourhood  by listing count of each neighbourhood group\n" + "--"*35)
print(Bottom_3_neighbourhoods_count)

````

````python
#problem statement-5
#which are most & least expensive neighbourhoods in each neighbourhood group ?((Bivariate + Bivariate Analysis))


# Group data by neighbourhood_group and neighbourhood, then calculate average room price
room_price_by_neighbourhood = airbnb.groupby(['neighbourhood_group', 'neighbourhood'])['price'].mean().round().reset_index()

# Sort the data by neighbourhood_group and average room price in descending order
room_price_by_neighbourhood = room_price_by_neighbourhood.sort_values(by=['neighbourhood_group', 'price'], ascending=[True, False])

# Select top 3 neighbourhoods within each neighbourhood_group
top_3_neighbourhoods_price = room_price_by_neighbourhood.groupby('neighbourhood_group').head(3)

#printing result
print (f"Top 3 neighbourhood by listing price of each neighbourhood group\n" + "--"*33)
print(top_3_neighbourhoods_price)

# Select bottom 3 neighborhoods within each neighbourhood_group
bottom_3_neighbourhoods_price = room_price_by_neighbourhood.groupby('neighbourhood_group').tail(3)

#printing result
print (f"\nBottom 3 neighbourhood by listing price of each neighbourhood group\n" + "--"*35)
print(bottom_3_neighbourhoods_price)

````

````python
#problem statement-6
# How does room type affect the minimum_nights requirement?(Bivariate Analysis)


#Group data by room_type ,minimum_nights
room_type_min_nights = airbnb.groupby('room_type')['minimum_nights'].median().astype("int")


#printing result
print (f"Room type & minimum night of stay\n" + "---"*11)
print(room_type_min_nights)
````

````python
#problem statement-7
# how room avalilablity  varies across room types in each neighbourhood group?(Bivariate analysis)

# Group data by neighbourhood_group, room_type, and calculate average availability
room_availablity = airbnb.groupby(['neighbourhood_group', 'room_type'])['availability_365'].mean().astype("int").reset_index()

#printing result
print (f"Average days of availablity of each roomtype in each neighbourhood group \n" + "---"*24)
print(room_availablity)
````

````python
#problem statement-8
#How does the number of reviews vary from different room types in each neighborhood group?(Bivariate Analysis)

# Group data by neighbourhood_group, room_type, and calculate average of total no.of reviews
room_type_reviews=airbnb.groupby(["neighbourhood_group","room_type"])["number_of_reviews"].mean().astype("int").reset_index()

#printing result
print (f"Average no.of reviews received by each roomtype in each neighbourhood group \n" + "---"*25)
print(room_type_reviews)
````

````python
#problem statement-9
#Which host receives the highest number of reviews?(Univariate Analysis)

# Group the data by  host name and calculate the total number of reviews of top 10
top_hosts_reviews = airbnb.groupby('host_name')['number_of_reviews'].sum().nlargest(10)

#printing result
print (f"Total number of reviews received by top 10 hosts \n" + "---"*16)
print(top_hosts_reviews)


````

````python
#problem statement-10
#Which host have the highest number of properties listed ?(top 10)(Univariate Analysis)


# Find the top 10 hosts with the highest number of properties
top_host_property = airbnb.groupby(['host_id', 'host_name'])['calculated_host_listings_count'].max().nlargest(10)

#printing result
print (f"Top 10 hosts in terms of property\n" + "---"*11)
print(top_host_property)
````

````python
#problem statement-11
#which are top 10 listed properties how their price & availablity varies ?(Multivariate Analysis)

# Group the data by property name and calculate the total count, average cost, and average availability
top_properties = airbnb.groupby('name').agg({'id': 'count', 'price': 'mean','availability_365': 'mean'}).reset_index()

# Rename the columns for clarity
top_properties.columns = ['Property Name', 'Listing Count', 'Price', 'Availability']

# Round the Price &  Availablity
top_properties['Price'] = top_properties['Price'].round()
top_properties['Availability'] = top_properties['Availability'].round()

# Sort the properties by listing count in descending order
top_properties = top_properties.sort_values(by='Listing Count', ascending=False)

# Get the top 10 properties
top_10_properties = top_properties.head(10)

#printing result
print (f"Top 10 listed propertied with price & availablity \n" + "---"*15)
top_10_properties

````

````python
#problem statement-12
# Is there any corelation & pattern between all quantitative values?(Multivariate Analysis)

#Exclude id,host_id
updated_airbnb=airbnb.drop(["id","host_id"],axis=1)

#find the co-relation metrix
corelation=updated_airbnb.corr()

#printing result
print (f"Co-relation metrix\n" + "---"*7)
corelation
````

The pair plot visualization is chosen for its ability to showcase the correlations and patterns between pairs of numerical variables in a comprehensive manner. It allows us to simultaneously examine the distributions of individual variables along with their potential relationships with each other.

By using a pair plot with kernel density estimate (KDE) plots on the diagonal, we can observe both the distributions of the variables and the scatter plots of variable pairs. This helps to identify potential linear relationships, clusters, and outliers.
